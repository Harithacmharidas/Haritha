# Pre-Processing

# Missing Value Handling 

```python
data.isna().sum()
```

```python
data['weight'] = data['weight'].str.replace('lbs', '')
data['weight'] = pd.to_numeric(data['weight'], errors='coerce')
```

```python
num_with_na=['weight', 'ball_controlling_skills', 'jumping_skills', 'penalties_conversion_rate', 'mental_strength',
            'shot_accuracy','behaviour_rating', 'matches_played', 'fitness_rating','years_of_experience' ]
for i in num_with_na:
    data[i]=data[i].fillna(data[i].median())
```

```python
cat_with_na = ['gender','strong_foot','coaching']
for i in cat_with_na:
    data[i]=data[i].fillna(data[i].mode()[0])
```

# Modelling

```python
#Here target variable is selection. so we are going for classification problem
```

```python
x=data.drop(["selection","name","player_id","country","strong_foot","height"],axis=1)
```

```python
y=data["selection"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred_lr=logit_model.predict(x_test)


import warnings
warnings.filterwarnings("ignore",category=UserWarning)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
print("Accuracy=",accuracy_score(y_test,y_pred_lr))
print("Precision=",precision_score(y_test,y_pred_lr,average="macro"))
print("recall_score=",recall_score(y_test,y_pred_lr,average="macro"))
print("f1_score=",f1_score(y_test,y_pred_lr,average="macro"))
```

We can look for different techniques and then decide based on these measures

# K-NN

```python
from sklearn.neighbors import KNeighborsClassifier
metric = []
neighbors = np.arange(3,15)

for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski',p=2)
    classifier.fit(x_train,y_train)
    y_pred_knn = classifier.predict(x_test)
    acc = accuracy_score(y_test,y_pred_knn)
    metric.append(acc)
```

```python
plt.plot(neighbors,metric,"o-")
plt.xlabel("kvalue")
plt.xlabel("accuracy")
plt.grid()
```

```python
classifier = KNeighborsClassifier(n_neighbors = 12, metric = 'minkowski',p=2)
classifier.fit(x_train,y_train)
y_pred_knn = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print("Accuracy=",accuracy_score(y_test,y_pred_knn))
print("Precision=",precision_score(y_test,y_pred_knn,average="macro"))
print("recall_score=",recall_score(y_test,y_pred_knn,average="macro"))
print("f1_score=",f1_score(y_test,y_pred_knn,average="macro"))
```

By this method, we can find the best model and fine tune and then we can classification problem
